```python
import zipfile

from PIL import Image
import pytesseract
import cv2 as cv
import numpy as np

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

```python
def findFaces(pil_img, scale=1.05):
    img = np.array(pil_img)
    img = img[:, :, ::-1].copy()

    # And we'll convert it to grayscale using the cvtColor image
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    # The next step is to use the face_cascade classifier. I'll let you go explore the docs if you
    # would like to, but the norm is to use the detectMultiScale() function. This function returns
    # a list of objects as rectangles. The first parameter is an ndarray of the image.
    faces = face_cascade.detectMultiScale(gray, scaleFactor=scale, minNeighbors=5)
    # And lets just print those faces out to the screen
    return faces
```

```python
def onlyFaces(pil_img, faces):
    face_list = list()
    for x, y, w, h in faces:
        only_face = pil_img.crop((x, y, x+w, y+h))
        max_size = (100, 100)
        only_face.thumbnail(max_size)
        face_list.append(only_face)
    return face_list
```

```python
def printToContactSheet(only_faces):
    if len(only_faces)%5 == 0:
        height = len(only_faces)/5
    else:
        height = (len(only_faces)/5)+1
    
    first_image = only_faces[0]
    contact_sheet=Image.new(first_image.mode, (100*5, 100*int(height)))
    x,y = (0,0)
    for face in only_faces:
        contact_sheet.paste(face, (x, y))            
        if x+100 == contact_sheet.width:
            x=0
            y=y+100
        else:
            x=x+100
    return contact_sheet
```

```python

def getRidofOverlapping(faces):
    removeList = list()
    for i in range(len(faces)-1):
        face1 = faces[i]
        face2 = faces[i+1]
        face1topLeft = (face1[0], face1[1])
        face1bottomRight = (face1[0]+face1[2], face1[1]+face1[3])

        face2topLeft = (face2[0], face2[1])
        face2bottomRight = (face2[0]+face2[2], face2[1]+face2[3])

        # Checking top left corner
        if face2topLeft[0] >= face1topLeft[0] and face2topLeft[0] <= face1bottomRight[0] and face2topLeft[1] >= face1topLeft[1] and face2topLeft[1] <= face1bottomRight[1]:
#             print('Overlapping')
            if face1[3] >= face2[3]:
                removeList.append(i+1)
            else:
                removeList.append(i)

        # Checking top right corner

        elif face2bottomRight[0] >= face1topLeft[0] and face2bottomRight[0] <= face1bottomRight[0] and face2topLeft[1] >= face1topLeft[1] and face2topLeft[1] <= face1bottomRight[1]:
#             print('Overlapping')
            if face1[3] >= face2[3]:
                removeList.append(i+1)
            else:
                removeList.append(i)

        # Checking bottomLeft corner

        elif face2topLeft[0] >= face1topLeft[0] and face2topLeft[0] <= face1bottomRight[0] and face2bottomRight[1] >= face1topLeft[1] and face2bottomRight[1] <= face1bottomRight[1]:
#             print('Overlapping')
            if face1[3] >= face2[3]:
                removeList.append(i+1)
            else:
                removeList.append(i)


        # Checking bottomRight corner

        elif face2bottomRight[0] >= face1topLeft[0] and face2bottomRight[0] <= face1bottomRight[0] and face1bottomRight[1] >= face1topLeft[1] and face2bottomRight[1] <= face1bottomRight[1]:
#             print('Overlapping')
            if face1[3] >= face2[3]:
                removeList.append(i+1)
            else:
                removeList.append(i)
        
    # Out of for loop
#     print(removeList)
    new_faces = faces
    for num in removeList:
        new_faces = np.delete(new_faces, num, axis=0)
    return new_faces
```

```python
def sortByX(faces):
    new_faces = faces

    for _ in range(len(new_faces)):    
        for i in range(-1, -1 * (len(new_faces)), -1):
            if new_faces[i][0] < new_faces[i-1][0]:
                new_faces[[i, i-1]] = new_faces[[i-1, i]]
    return new_faces    
```

```python
def binarize(image_to_transform, threshold=180):
    output_image=image_to_transform.convert("L")
    for x in range(output_image.width):
        for y in range(output_image.height):
            if output_image.getpixel((x,y))< threshold:
                output_image.putpixel( (x,y), 0 )
            else:
                output_image.putpixel( (x,y), 255 )
    return output_image
```

```python
with zipfile.ZipFile("readonly/images.zip") as z:
    files = z.infolist()

result_list = list()
for file in files:
    my_D = dict()
    my_D["file name"] = file.filename
    image = zipfile.ZipFile("readonly/images.zip").open(my_D["file name"])
    my_D["image"] =  Image.open(image).convert("RGB")
    faces = findFaces(my_D["image"], scale=1.3)
    if len(faces)>0: 
        # I'm sorting face boxes by x1 value for detecting overlapping
        faces = sortByX(faces)
        # With this function I'm removing one of the overlapping rectangles
        faces = getRidofOverlapping(faces)
        my_D["bounding boxes"] = faces
        # Extracting only faces from whole page
        only_faces = onlyFaces(my_D["image"], faces)
        # Printing the faces to a contact sheet
        my_D["contact sheet"] = printToContactSheet(only_faces)
    else:
        # If there is no face in the page I'm setting values below to false
        my_D["bounding boxes"] = False
        my_D["contact sheet"] = False
    
    # For extracting the text I'm binarizing the image
    text_img = binarize(my_D["image"])
    # taking the text from image
    text = pytesseract.image_to_string(text_img)
    my_D["text on page"] = text
    # And finally I'm addind the dictionary to my list
    result_list.append(my_D)
```

```python
name = input("Name: ")

for dictionary in result_list:
    if name in dictionary["text on page"]:
        print("Results found in file {}".format(dictionary["file name"]))        
        if dictionary["contact sheet"]:
            display(dictionary["contact sheet"])
        else:
            print("But there were no faces in that file!")
```
